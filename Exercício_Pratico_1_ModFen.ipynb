# **Exercício 1)**

Temos a fração abaixo, e um limite que tende a 3:

$$f(x) = \frac{x^2-4}{x - 2}$$

Logo, para termos os valores da aproximação pela direita e pela esquerda substituímos valores próximos de 3.


$$f(2.9) = \frac{2.9^2-4}{2.9 - 2} = 3,51 $$

$$f(2.99) = \frac{2.99^2-4}{2.99 - 2} = 3,9501$$

$$f(2.999) = \frac{2.999^2-4}{2.999 - 2} = 3,995001$$

$$f(3.001) = \frac{3.001^2-4}{3.001 - 2} = 5.1$$

$$f(3.01) = \frac{3.01^2-4}{3.01 - 2} = 5.01$$

$$f(3.1) = \frac{3.1^2-4}{3.1 - 2} = 5.001$$



```python
def f(x):
    return x**2-4/x-2

c = 3
L = np.inf

x = np.linspace(-3, 3, 100)
y = f(x)

plt.plot(x, y, label='x**2-4/x-2')
plt.axvline(x=c, color='gray', linestyle='--', label=f'x = {c}')
plt.axhline(y=L, color='gray', linestyle='--', label=f'y = {L}')
plt.legend()
plt.show()
```

$$f(x) = \frac{x^2-3x+2}{x - 1}$$

Como é impossível de obter o resultado, manipula-se a função para que seja possível realizar o cálculo do limite tendendo a 1:


$$f(x) = \frac{(x-2)(x-1)}{x - 1} = -1$$

$$f(9) =  \frac{\sqrt{9}-3}{9-9} = \frac{0}{0}$$

$$g(-3) =  -3^2+3 = 12, x>=3$$

# **Exercício 4**
Equação 1:
$$\frac{x-3}{7-x}$$

Para que esse limite não seja indefinido, o 'x' não pode tender a 7. Senão o limite será indefinido

Equação 2:
$$\frac{x-1}{x^2-5x-6}$$

Para que esse limite não seja indefinido, o x precisa ser diferente de -1 e diferente de 6. Senão o limite será indefinido.

Equação 3:
$$\frac{1}{x^2+5x}$$

Para que esse limite não seja indefinido, o x precisa ser diferente de 0. Senão o limite será indefinido.

Equação 3:
$$\frac{3+x}{x^2-2}$$

Para que esse limite não seja indefinido, o x precisa ser diferente de: $$\sqrt{2}$$ Senão o limite será indefinido.

```python
#Gráfico 1

def f(x):
    return (x-3)/(7-x)

c = 7
L = np.inf

x = np.linspace(-7, 30, 1000)
y = f(x)

plt.plot(x, y, label='Gráfico 1  Ex 1')
plt.axvline(x=c, color='gray', linestyle='--', label=f'x = {c}')
plt.axhline(y=L, color='gray', linestyle='--', label=f'y = {L}')
plt.legend()
plt.show()

#Gráfico 2

def f(x):
    return (x-1)/(x**2-5*x-6)

c = -1
L = np.inf

x = np.linspace(-20, 20, 1000)
y = f(x)

plt.plot(x, y, label='Gráfico 1  Ex 2')
plt.axvline(x=c, color='gray', linestyle='--', label=f'x = {c}')
plt.axhline(y=L, color='gray', linestyle='--', label=f'y = {L}')
plt.legend()
plt.show()

#Gráfico 3

def f(x):
    return (x-1)/(x**2+5*x)

c = 0
L = np.inf

x = np.linspace(-3, 5, 1000)
y = f(x)

plt.plot(x, y, label='Gráfico 1  Ex 3')
plt.axvline(x=c, color='gray', linestyle='--', label=f'x = {c}')
plt.axhline(y=L, color='gray', linestyle='--', label=f'y = {L}')
plt.legend()
plt.show()


#Gráfico 4

def f(x):
    return (3+x)/(x**2-2)

c = np.sqrt(2)
L = np.inf

x = np.linspace(-2, 2, 1000)
y = f(x)

plt.plot(x, y, label='Gráfico 1  Ex 3')
plt.axvline(x=c, color='gray', linestyle='--', label=f'x = {c}')
plt.axhline(y=L, color='gray', linestyle='--', label=f'y = {L}')
plt.legend()
plt.show()
```
